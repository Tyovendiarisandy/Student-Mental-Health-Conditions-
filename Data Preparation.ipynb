##**4. Data Formatting**

```python
# delete columns that are not used in the analysis
df= df.drop(columns='Timestamp', axis=1)
```

```python
# rename the column to ease the code process
df = df.rename(columns={
                  'Choose your gender':'gender',
                  'Age':'age',
                  'What is your course?':'study_programs',
                  'Your current year of Study':'study_years',
                  'What is your CGPA?':'cgpa',
                  'Marital status':'marital_status',
                  'Do you have Depression?':'depression',
                  'Do you have Anxiety?':'anxiety',
                  'Do you have Panic attack?':'panic_attack',
                  'Did you seek any specialist for a treatment?':'seeking_treatment'})
```

```python
# change the data type of age column
df['age']= df['age'].astype('int64')

# change all data in rows to lowercase format
df = df.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
```

```python
# Membuat kondisi untuk menentukan nilai pada kolom "mental_health_issues"
condition = ((df['depression'] == 'yes') | (df['anxiety'] == 'yes') | (df['panic_attack'] == 'yes'))

# Menggunakan np.where untuk memberikan nilai "yes" atau "no" berdasarkan kondisi
df['mental_health_issues'] = np.where(condition, 'yes', 'no')
```

```python
# Menggunakan kondisi untuk menentukan nilai pada kolom "mental_health_issues"
conditions = [
    (df['depression'] == 'yes') & (df['anxiety'] == 'no') & (df['panic_attack'] == 'no'),
    (df['depression'] == 'no') & (df['anxiety'] == 'yes') & (df['panic_attack'] == 'no'),
    (df['depression'] == 'no') & (df['anxiety'] == 'no') & (df['panic_attack'] == 'yes'),
    (df['depression'] == 'yes') & (df['anxiety'] == 'yes') & (df['panic_attack'] == 'no'),
    (df['depression'] == 'yes') & (df['anxiety'] == 'no') & (df['panic_attack'] == 'yes'),
    (df['depression'] == 'no') & (df['anxiety'] == 'yes') & (df['panic_attack'] == 'yes'),
    (df['depression'] == 'yes') & (df['anxiety'] == 'yes') & (df['panic_attack'] == 'yes')
]

# Menentukan kategorisasi dari kondisi sebelumnya
choices = ['mild', 'mild', 'mild', 'moderate', 'moderate', 'moderate', 'severe']

# Menggunakan np.select untuk mengisi baris pada kolom "mental_health_issues"
df['mental_health_category'] = np.select(conditions, choices, default='normal')
```

```python
df.info()
```

```python
# categorizes the values in the list if any similarity in field/course
a = ['fiqh', 'usuluddin ', 'fiqh fatwa ', 'kirkhs', 'irkhs']
b = ['engine', 'engin']
c = ['econs', 'kenms', 'enm']
d = ['pendidikan islam ', 'pendidikan islam', 'islamic education', 'koe', 'benl', 'diploma tesl', 'taasl']
e = ['human resources', 'mhsc', 'ala']
f = ['bit', 'it']

# using function to perform value replacement
def replace_multiple_values(value):
    if value in a:
        return 'irkhs'
    elif value in b:
        return 'engineering'
    elif value in c:
        return 'economics'
    elif value in d:
        return 'education'
    elif value in e:
        return 'human_sciences'
    elif value in f:
        return 'information_tecnology'
    else:
        return value

# using function apply() with replace_multiple_values
df['study_programs'] = df['study_programs'].apply(replace_multiple_values)
```

```python
# using .replace to change older value to the new one
df['study_programs'].replace({'bcs':'computer_science',
                              'laws':'law',
                              'malcom':'communication',
                              'communication ':'communication',
                              'accounting ':'accounting',
                              'mathemathics': 'mathematics',
                              'kop': 'pharmacy',
                              'diploma nursing': 'nursing',
                              'nursing ':'nursing'
                              }, inplace=True)
```

```python
# provides an underline separator for each data with a space
df['study_programs'] = df['study_programs'].apply(lambda x: x.replace(' ', '_'))

# provides an underline separator for each data with a space
df['study_years'] = df['study_years'].apply(lambda x: x.replace(' ', '_'))

# remove the space behind the word in the column 'study programs'
df['study_programs'] = df['study_programs'].str.rstrip('_')

# remove the space behind the word in the column 'cgpa'
df['cgpa'] = df['cgpa'].str.rstrip(' ')
```
